```python
# ================================================================
#  FMCG CONSUMER TREND PREDICTION — HYBRID MACHINE LEARNING MODEL
#  Context  : Hindustan Unilever Limited (HUL)
#  Submitted: IT Applications | IIM Rohtak
#  Pipeline : Data Ingestion → Preprocessing → Unsupervised
#             Clustering → Supervised ML → Evaluation → Insights
# ================================================================

# ================================
# SECTION 1 — IMPORT LIBRARIES
# ================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

from sklearn.preprocessing import StandardScaler, label_binarize
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.ensemble import (RandomForestClassifier, RandomForestRegressor,
                               GradientBoostingClassifier)
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import (
    r2_score, mean_squared_error,
    f1_score, accuracy_score, precision_score, recall_score,
    classification_report, confusion_matrix
)
from scipy.stats import f_oneway

plt.rcParams["figure.dpi"] = 120
plt.rcParams["font.family"] = "DejaVu Sans"

PRODUCTS   = ["facecream","facewash","toothpaste","bathingsoap","shampoo","moisturizer"]
PROD_NAMES = ["Face Cream","Face Wash","Toothpaste","Bathing Soap","Shampoo","Moisturizer"]
COLORS     = ["#1f77b4","#ff7f0e","#2ca02c","#d62728","#9467bd","#8c564b"]
C_PALETTE  = ["#264653","#2a9d8f","#e76f51"]

print("All libraries imported successfully.")
print("="*60)
print("   FMCG TREND PREDICTION — HUL HYBRID ML PIPELINE")
print("="*60)

# ================================
# SECTION 2 — LOAD DATA
# ================================
from google.colab import files
uploaded = files.upload()          # ← Upload your CSV when prompted

import io
filename = list(uploaded.keys())[0]
df = pd.read_csv(io.BytesIO(uploaded[filename]))

# Feature Engineering
df["profit_per_unit"]    = df["total_profit"] / df["total_units"]
df["month_cycle"]        = ((df["month_number"] - 1) % 12) + 1
df["year"]               = ((df["month_number"] - 1) // 12) + 1
df["sales_growth_rate"]  = df["total_units"].pct_change().fillna(0)
df["high_margin_month"]  = (df["profit_per_unit"] > df["profit_per_unit"].median()).astype(int)

print(f"\nDataset Shape: {df.shape}")
print(df.head())

# ================================
# SECTION 3 — KEEP NUMERIC DATA
# ================================
numeric_df = df.select_dtypes(include=[np.number]).dropna().copy()

print("\nNumeric Columns:")
print(list(numeric_df.columns))
print(f"\nDescriptive Statistics:")
print(numeric_df[PRODUCTS + ['total_units','total_profit']].describe().round(2))

# ================================
# SECTION 4 — DISTRIBUTION
#  — Understand sales spread across
#    months for trend detection
# ================================
print("\n" + "─"*50)
print("  SECTION 4: DISTRIBUTION ANALYSIS")
print("─"*50)

fig, axes = plt.subplots(2, 3, figsize=(15, 8))
axes = axes.flatten()

for i, (prod, name, color) in enumerate(zip(PRODUCTS, PROD_NAMES, COLORS)):
    axes[i].hist(df[prod], bins=20, color=color, edgecolor="black", alpha=0.85)
    axes[i].axvline(df[prod].mean(),   color="red",    linestyle="--",
                    linewidth=1.8, label=f"Mean: {df[prod].mean():,.0f}")
    axes[i].axvline(df[prod].median(), color="orange", linestyle="--",
                    linewidth=1.8, label=f"Median: {df[prod].median():,.0f}")
    axes[i].set_title(f"Distribution of {name}", fontweight="bold")
    axes[i].set_xlabel("Units Sold")
    axes[i].set_ylabel("Frequency")
    axes[i].legend(fontsize=7)
    axes[i].grid(axis="y", alpha=0.3)

fig.suptitle("FMCG Product Sales Distribution — HUL\n"
             "Identifying Sales Spread & Emerging Trend Patterns",
             fontsize=13, fontweight="bold")
plt.tight_layout()
plt.savefig("01_distribution.png", bbox_inches="tight")
plt.show()

# Distribution of Total Profit (main target)
plt.figure(figsize=(8, 4))
sns.histplot(df["total_profit"], kde=True, color="#264653", bins=25)
plt.axvline(df["total_profit"].mean(),   color="red",    linestyle="--",
            linewidth=2, label=f"Mean:   ₹{df['total_profit'].mean():,.0f}")
plt.axvline(df["total_profit"].median(), color="orange", linestyle="--",
            linewidth=2, label=f"Median: ₹{df['total_profit'].median():,.0f}")
plt.title("Distribution of Total Profit — HUL FMCG", fontweight="bold")
plt.xlabel("Total Profit (₹)"); plt.ylabel("Frequency")
plt.legend(); plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig("02_profit_distribution.png", bbox_inches="tight")
plt.show()
print("Distribution analysis complete.")

# ================================
# SECTION 5 — CORRELATION MATRIX
#  — Which products drive profit?
#  — Core to FMCG trend detection
# ================================
print("\n" + "─"*50)
print("  SECTION 5: CORRELATION MATRIX")
print("─"*50)

corr_cols = PRODUCTS + ["total_units", "total_profit", "profit_per_unit"]
corr      = df[corr_cols].corr()

plt.figure(figsize=(11, 7))
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True, fmt=".2f", cmap="coolwarm",
            center=0, linewidths=0.5, annot_kws={"size": 9}, vmin=-1, vmax=1)
plt.title("🔗 Correlation Matrix — FMCG Product Sales vs Profit\n"
          "(Hindustan Unilever | Identifying Key Revenue Drivers)",
          fontsize=12, fontweight="bold", pad=15)
plt.tight_layout()
plt.savefig("03_correlation_matrix.png", bbox_inches="tight")
plt.show()

print("\nTop Correlations with Total Profit:")
corr_profit = corr["total_profit"].drop("total_profit").sort_values(ascending=False)
for feat, val in corr_profit.items():
    bar = "█" * int(abs(val) * 20)
    sign = "+" if val > 0 else "-"
    print(f"  {feat:<20}  r = {val:+.4f}  {sign}{bar}")
print("Correlation matrix complete.")

# ================================
# SECTION 6 — FEATURE SCALING
# ================================
scaler      = StandardScaler()
scaled_data = scaler.fit_transform(df[PRODUCTS])
print("\nFeatures scaled with StandardScaler.")

# ================================
# SECTION 7 — K-MEANS CLUSTERING
#  STEP 2 of Hybrid Model
#  Unsupervised: Discover hidden
#  consumer behavior segments
# ================================
print("\n" + "─"*50)
print("  SECTION 7: CLUSTER ANALYSIS (K-MEANS)")
print("─"*50)

# Elbow Method
inertias = []
for k in range(2, 9):
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(scaled_data)
    inertias.append(km.inertia_)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].plot(range(2, 9), inertias, "bo-", linewidth=2.5, markersize=8)
axes[0].axvline(x=3, color="red", linestyle="--", alpha=0.8, label="Optimal K = 3")
axes[0].set_xlabel("Number of Clusters (K)")
axes[0].set_ylabel("Inertia (Within-Cluster SS)")
axes[0].set_title("Elbow Method — Optimal K Selection\n(FMCG Consumer Demand Segments)",
                  fontweight="bold")
axes[0].legend(); axes[0].grid(alpha=0.3)

# Final KMeans with K=3
kmeans          = KMeans(n_clusters=3, random_state=42, n_init=10)
df["Cluster"]   = kmeans.fit_predict(scaled_data)

# Label clusters by profit level (meaningful for FMCG)
mean_profit     = df.groupby("Cluster")["total_profit"].mean()
high_c          = mean_profit.idxmax()
low_c           = mean_profit.idxmin()
mid_c           = [c for c in [0,1,2] if c not in [high_c, low_c]][0]
cluster_map     = {high_c: "High-Demand Months",
                   mid_c:  "Mid-Demand Months",
                   low_c:  "Low-Demand Months"}
df["Segment"]   = df["Cluster"].map(cluster_map)

print("\nCluster Distribution:")
print(df["Segment"].value_counts())

# PCA 2D Visualization
pca   = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(scaled_data)

for cid, cname in cluster_map.items():
    mask = df["Cluster"] == cid
    axes[1].scatter(X_pca[mask.values, 0], X_pca[mask.values, 1],
                    label=cname, alpha=0.7, s=55,
                    color=C_PALETTE[list(cluster_map.keys()).index(cid)],
                    edgecolors="black", linewidths=0.3)
axes[1].set_xlabel(f"PC1 ({pca.explained_variance_ratio_[0]*100:.1f}% variance)")
axes[1].set_ylabel(f"PC2 ({pca.explained_variance_ratio_[1]*100:.1f}% variance)")
axes[1].set_title("K-Means Clusters — PCA 2D View\n(Consumer Demand Segmentation)",
                  fontweight="bold")
axes[1].legend(fontsize=9); axes[1].grid(alpha=0.3)

fig.suptitle("Cluster Analysis — FMCG Consumer Demand Segments (HUL)",
             fontsize=13, fontweight="bold")
plt.tight_layout()
plt.savefig("04_cluster_analysis.png", bbox_inches="tight")
plt.show()

# Cluster scatter: BathingSoap vs Profit (highest volume product)
plt.figure(figsize=(8, 5))
for cid, cname in cluster_map.items():
    mask = df["Cluster"] == cid
    plt.scatter(df.loc[mask, "bathingsoap"], df.loc[mask, "total_profit"],
                label=cname, alpha=0.75, s=55,
                color=C_PALETTE[list(cluster_map.keys()).index(cid)],
                edgecolors="black", linewidths=0.3)
plt.xlabel("Bathing Soap Units Sold (Highest-Volume Product)")
plt.ylabel("Total Profit (₹)")
plt.title("Cluster Analysis — Bathing Soap vs Total Profit\n"
          "(K-Means | Unsupervised Learning | HUL FMCG)", fontweight="bold")
plt.legend(); plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig("04b_cluster_scatter.png", bbox_inches="tight")
plt.show()

# Cluster profile table
print("\nCluster Profile (Mean Values per Segment):")
profile = df.groupby("Segment")[PRODUCTS + ["total_profit","total_units"]].mean().round(0)
print(profile.to_string())

# ANOVA F-Score across clusters
print("\n" + "─"*50)
print("  ANOVA F-TEST — Feature Significance Across Clusters")
print("─"*50)
f_data = {}
for feat in PRODUCTS:
    groups = [df.loc[df["Cluster"]==c, feat].values for c in range(3)]
    f_val, p_val = f_oneway(*groups)
    sig = "Significant" if p_val < 0.05 else " Not Significant"
    f_data[feat] = {"F-Score": round(f_val, 2), "p-value": round(p_val, 4)}
    print(f"  {feat:<15}  F = {f_val:7.2f}  p = {p_val:.4f}  {sig}")

fscore_df = pd.DataFrame(f_data).T.sort_values("F-Score", ascending=True)
plt.figure(figsize=(9, 5))
bar_colors = ["#2ca02c" if p < 0.05 else "#d62728" for p in fscore_df["p-value"]]
fscore_df["F-Score"].plot(kind="barh", color=bar_colors, edgecolor="black", alpha=0.85)
plt.xlabel("ANOVA F-Score")
plt.title("ANOVA F-Score — Product Differentiation Across Demand Segments\n"
          "(Green = Statistically Significant  p < 0.05)", fontweight="bold")
plt.grid(axis="x", alpha=0.3)
plt.tight_layout()
plt.savefig("05_anova_fscores.png", bbox_inches="tight")
plt.show()
print("Cluster analysis complete.")

# ================================
# SECTION 8 — SUPERVISED ML MODEL
#  STEP 3 of Hybrid Model
#  Cluster label used as feature
#  (The "Hybrid" step)
#  Target: total_profit category
# ================================
print("\n" + "─"*50)
print("  SECTION 8: SUPERVISED LEARNING — HYBRID MODEL")
print("─"*50)

# Create 3-class target: Low / Medium / High profit months
df["profit_class"] = pd.qcut(df["total_profit"], q=3,
                              labels=["Low Profit","Medium Profit","High Profit"])

# Features = all products + total_units + Cluster (HYBRID step)
feature_cols = PRODUCTS + ["total_units", "Cluster"]
X = df[feature_cols]
y = df["profit_class"]

print("\nTarget Distribution (Profit Classes):")
print(y.value_counts())

# Train-Test Split (75/25 stratified — same as FoodLink methodology)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)
print(f"\nTrain: {X_train.shape[0]} samples | Test: {X_test.shape[0]} samples")

# ================================
# SECTION 9 — 8 CLASSIFIERS
#  Accuracy, F1, Precision, Recall
# ================================
print("\n" + "─"*50)
print("  SECTION 9: 8 CLASSIFIERS — MODEL COMPARISON")
print("─"*50)

classifiers = {
    "Random Forest"      : RandomForestClassifier(n_estimators=200, random_state=42),
    "Gradient Boosting"  : GradientBoostingClassifier(n_estimators=200, random_state=42),
    "SVM"                : SVC(kernel="rbf", probability=True, random_state=42),
    "Logistic Regression": LogisticRegression(max_iter=500, random_state=42),
    "Decision Tree"      : DecisionTreeClassifier(random_state=42),
    "KNN"                : KNeighborsClassifier(n_neighbors=5),
    "Naive Bayes"        : None,    # placeholder — replaced below
    "Extra Trees"        : None,
}

from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import ExtraTreesClassifier
classifiers["Naive Bayes"]  = GaussianNB()
classifiers["Extra Trees"]  = ExtraTreesClassifier(n_estimators=200, random_state=42)

results = {}
print(f"\n{'Model':<22} {'Accuracy':>9} {'F1':>9} {'Precision':>10} {'Recall':>8}")
print("─"*62)

for name, clf in classifiers.items():
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    acc  = accuracy_score(y_test, y_pred)
    f1   = f1_score(y_test, y_pred, average="weighted")
    prec = precision_score(y_test, y_pred, average="weighted", zero_division=0)
    rec  = recall_score(y_test, y_pred, average="weighted")
    results[name] = {"clf":clf, "y_pred":y_pred,
                     "accuracy":acc, "f1":f1, "precision":prec, "recall":rec}
    print(f"  {name:<20} {acc:>9.3f} {f1:>9.3f} {prec:>10.3f} {rec:>8.3f}")

# ─── Model Accuracy Comparison Chart (like FoodLink Fig 3) ───
model_names = list(results.keys())
accs  = [results[n]["accuracy"]  for n in model_names]
f1s   = [results[n]["f1"]        for n in model_names]
precs = [results[n]["precision"] for n in model_names]
recs  = [results[n]["recall"]    for n in model_names]

x     = np.arange(len(model_names))
width = 0.22

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Left: Accuracy bar
bar_colors = ["#b5835a" if a == max(accs) else "#2a9d8f" for a in accs]
bars = axes[0].bar(x, accs, color=bar_colors, edgecolor="black", width=0.6, alpha=0.9)
axes[0].set_xticks(x)
axes[0].set_xticklabels(model_names, rotation=35, ha="right", fontsize=9)
axes[0].set_ylim(0.5, 1.05)
axes[0].set_ylabel("Accuracy")
axes[0].set_title("Model Accuracy Comparison", fontweight="bold")
axes[0].grid(axis="y", alpha=0.3)
for bar, val in zip(bars, accs):
    axes[0].text(bar.get_x()+bar.get_width()/2, val+0.005,
                 f"{val:.3f}", ha="center", va="bottom", fontsize=8, fontweight="bold")

# Right: Accuracy vs F1
axes[1].bar(x - width/2, accs, width, label="Accuracy", color="#264653", alpha=0.85, edgecolor="black")
axes[1].bar(x + width/2, f1s,  width, label="F1 Score", color="#2a9d8f", alpha=0.85, edgecolor="black")
axes[1].set_xticks(x)
axes[1].set_xticklabels(model_names, rotation=35, ha="right", fontsize=9)
axes[1].set_ylim(0.5, 1.1)
axes[1].set_title("Accuracy vs F1 Score", fontweight="bold")
axes[1].legend(); axes[1].grid(axis="y", alpha=0.3)

fig.suptitle("Supervised Learning — Model Performance Comparison\n"
             "FMCG Profit Category Prediction (HUL)",
             fontsize=13, fontweight="bold")
plt.tight_layout()
plt.savefig("06_model_accuracy_comparison.png", bbox_inches="tight")
plt.show()

# ─── Precision / Recall / F1 grouped bar ─────────────────────
fig, ax = plt.subplots(figsize=(14, 5))
ax.bar(x - width,   precs, width, label="Precision", color="#264653", alpha=0.85, edgecolor="black")
ax.bar(x,           f1s,   width, label="F1 Score",  color="#2a9d8f", alpha=0.85, edgecolor="black")
ax.bar(x + width,   recs,  width, label="Recall",    color="#e76f51", alpha=0.85, edgecolor="black")
ax.set_xticks(x); ax.set_xticklabels(model_names, rotation=35, ha="right", fontsize=9)
ax.set_ylim(0.5, 1.1)
ax.set_title("Precision / F1 Score / Recall — All 8 Classifiers\n(FMCG Demand Trend Classification | HUL)",
             fontweight="bold")
ax.legend(); ax.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.savefig("07_precision_f1_recall.png", bbox_inches="tight")
plt.show()
print("8-classifier comparison complete.")

# ================================
# SECTION 10 — MODEL EVALUATION
#  Best Model: R², RMSE, F1 Score
# ================================
print("\n" + "─"*50)
print("  SECTION 10: MODEL EVALUATION (R², RMSE, F1)")
print("─"*50)

best_name = max(results, key=lambda k: results[k]["accuracy"])
best_clf  = results[best_name]["clf"]
best_pred = results[best_name]["y_pred"]

# F1 Score (best classifier)
f1_best      = results[best_name]["f1"]
f1_macro     = f1_score(y_test, best_pred, average="macro")
f1_weighted  = f1_score(y_test, best_pred, average="weighted")

# R² and RMSE via regression (Random Forest on same features)
rf_reg   = RandomForestRegressor(n_estimators=200, random_state=42)
rf_reg.fit(X_train.copy(),
           df.loc[X_train.index, "total_profit"])
y_pred_reg = rf_reg.predict(X_test)
r2   = r2_score(df.loc[X_test.index, "total_profit"], y_pred_reg)
rmse = np.sqrt(mean_squared_error(df.loc[X_test.index, "total_profit"], y_pred_reg))

# Convert to 3-class for F-score consistency
y_reg_true  = df.loc[X_test.index, "total_profit"]
y_test_cat  = pd.qcut(y_reg_true, 3, labels=[0,1,2])
y_pred_cat  = pd.qcut(pd.Series(y_pred_reg), 3, labels=[0,1,2])
fscore_reg  = f1_score(y_test_cat, y_pred_cat, average="weighted")

print(f"\n===== MODEL RESULTS (Hybrid ML — HUL FMCG) =====")
print(f"Best Classifier   : {best_name}")
print(f"Accuracy          : {results[best_name]['accuracy']:.3f}")
print(f"F1 Score (Weighted): {f1_weighted:.3f}")
print(f"F1 Score (Macro)  : {f1_macro:.3f}")
print(f"Precision         : {results[best_name]['precision']:.3f}")
print(f"Recall            : {results[best_name]['recall']:.3f}")
print(f"─"*45)
print(f"R Square Score    : {r2:.3f}")
print(f"RMSE              : {rmse:.3f}")
print(f"F1 Score (Reg→Cat): {fscore_reg:.3f}")
print(f"="*45)

# Actual vs Predicted (Regression)
y_actual = df.loc[X_test.index, "total_profit"]
plt.figure(figsize=(7, 5))
plt.scatter(y_actual, y_pred_reg, alpha=0.65, color="#2a9d8f",
            edgecolors="black", s=50)
lim = [min(y_actual.min(), y_pred_reg.min()),
       max(y_actual.max(), y_pred_reg.max())]
plt.plot(lim, lim, "r--", linewidth=2, label="Perfect Fit")
plt.xlabel("Actual Total Profit (₹)")
plt.ylabel("Predicted Total Profit (₹)")
plt.title(f"Actual vs Predicted — Random Forest Regressor\n"
          f"R² = {r2:.3f}  |  RMSE = {rmse:,.1f}  |  F1 = {fscore_reg:.3f}",
          fontweight="bold")
plt.legend(); plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig("08_actual_vs_predicted.png", bbox_inches="tight")
plt.show()

# Confusion Matrix (Best Classifier)
plt.figure(figsize=(7, 5))
cm     = confusion_matrix(y_test, best_pred,
                          labels=["Low Profit","Medium Profit","High Profit"])
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Low","Medium","High"],
            yticklabels=["Low","Medium","High"])
plt.xlabel("Predicted"); plt.ylabel("Actual")
plt.title(f"Confusion Matrix — {best_name}\n"
          f"Accuracy = {results[best_name]['accuracy']:.3f}  "
          f"|  F1 = {f1_weighted:.3f}",
          fontweight="bold")
plt.tight_layout()
plt.savefig("09_confusion_matrix.png", bbox_inches="tight")
plt.show()

# Classification Report
print(f"\nClassification Report ({best_name}):")
print(classification_report(y_test, best_pred,
      target_names=["Low Profit","Medium Profit","High Profit"]))
print("Model evaluation complete.")

# ================================
# SECTION 11 — FEATURE IMPORTANCE
#  Which products most influence
#  profit prediction (FMCG insight)
# ================================
print("\n" + "─"*50)
print("  SECTION 11: FEATURE IMPORTANCE")
print("─"*50)

importance = pd.Series(rf_reg.feature_importances_,
                       index=feature_cols).sort_values(ascending=True)

plt.figure(figsize=(9, 5))
bar_imp = ["#d62728" if v >= importance.quantile(0.6) else "#1f77b4"
           for v in importance]
importance.plot(kind="barh", color=bar_imp, edgecolor="black", alpha=0.88)
plt.axvline(importance.mean(), color="orange", linestyle="--",
            linewidth=2, label=f"Mean = {importance.mean():.3f}")
plt.xlabel("Importance Score")
plt.title("Feature Importance — Key FMCG Profit Drivers\n"
          "(Random Forest | HUL Consumer Trend Prediction)",
          fontweight="bold")
plt.legend(); plt.grid(axis="x", alpha=0.3)
plt.tight_layout()
plt.savefig("10_feature_importance.png", bbox_inches="tight")
plt.show()

print("\nFeature Importance Ranking:")
for feat, val in importance.sort_values(ascending=False).items():
    bar = "█" * int(val * 80)
    print(f"  {feat:<15}  {val:.4f}  {bar}")
print("Feature importance complete.")

# ================================
# SECTION 12 — BROWNIE POINTS
#  Monthly Sales Trend &
#  Segment Profit Bar Chart
#  (Extra strategic insights)
# ================================
print("\n" + "─"*50)
print("  SECTION 12: STRATEGIC INSIGHTS (BONUS)")
print("─"*50)

# A) Product-wise Sales Trend Over Time
fig, ax = plt.subplots(figsize=(13, 5))
for prod, name, color in zip(PRODUCTS, PROD_NAMES, COLORS):
    ax.plot(df["month_number"], df[prod],
            label=name, linewidth=1.8, color=color, alpha=0.85)
ax.set_xlabel("Month Number")
ax.set_ylabel("Units Sold")
ax.set_title("Product-wise Sales Trend — 120 Months (HUL FMCG)\n"
             "Identifying Emerging Consumer Preferences", fontweight="bold")
ax.legend(ncol=3, fontsize=8); ax.grid(alpha=0.3)
plt.tight_layout()
plt.savefig("11_sales_trend.png", bbox_inches="tight")
plt.show()

# B) Average Profit by Consumer Demand Segment
seg_profit = df.groupby("Segment")["total_profit"].mean().sort_values()
plt.figure(figsize=(7, 4))
bars = plt.barh(seg_profit.index, seg_profit.values,
                color=C_PALETTE, edgecolor="black", alpha=0.9)
for bar, val in zip(bars, seg_profit.values):
    plt.text(val + 1000, bar.get_y()+bar.get_height()/2,
             f"₹{val:,.0f}", va="center", fontsize=9, fontweight="bold")
plt.xlabel("Average Total Profit (₹)")
plt.title("Avg Profit by Consumer Demand Segment\n"
          "(K-Means Cluster Output | HUL FMCG)", fontweight="bold")
plt.grid(axis="x", alpha=0.3)
plt.tight_layout()
plt.savefig("12_segment_profit.png", bbox_inches="tight")
plt.show()

# C) Cross-Validation (5-Fold) for Best Classifier
cv_scores = cross_val_score(best_clf, X, y, cv=5, scoring="f1_weighted")
print(f"\n5-Fold Cross-Validation ({best_name}):")
print(f"  CV F1 Scores : {[round(s,3) for s in cv_scores]}")
print(f"  Mean F1      : {cv_scores.mean():.3f}")
print(f"  Std Dev      : {cv_scores.std():.3f}")

# ================================
# FINAL SUMMARY TABLE
# ================================
print("\n" + "="*60)
print("       FINAL RESULTS SUMMARY — HUL FMCG HYBRID ML")
print("="*60)
print(f"  Dataset         : {len(df)} months | 6 HUL products")
print(f"  Classifiers     : 8 supervised ML models")
print(f"  Best Model      : {best_name}")
print(f"  Accuracy        : {results[best_name]['accuracy']:.3f}")
print(f"  F1 Score        : {f1_weighted:.3f}")
print(f"  Precision       : {results[best_name]['precision']:.3f}")
print(f"  Recall          : {results[best_name]['recall']:.3f}")
print(f"  R² Score        : {r2:.3f}")
print(f"  RMSE            : ₹{rmse:,.1f}")
print(f"  Clusters Found  : 3 (High / Mid / Low Demand)")
print(f"  Top Profit Driver: {importance.idxmax()}")
print(f"  CV F1 (5-fold)  : {cv_scores.mean():.3f} ± {cv_scores.std():.3f}")
print("="*60)
print("\nAll 12 charts saved. Download from Files panel (left sidebar).")
```
